```python
import numpy as np
#import matplotlib_inline as plt
#import matplotlib.image as pltimg
import matplotlib.pyplot as plt
import cv2
```

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
```

# <b>1.  Edge Detection

### Read and display penguin_image

```python
penguin_image = plt.imread('penguin.jpg')
show_image(penguin_image, 'Original Penduin Image')
```

### Use Scikit-image to display the gray scale of the penguin image

```python
from skimage import color
```

```python
gray_penguin_image = color.rgb2gray(penguin_image)
show_image(gray_penguin_image, 'Original Penduin Image')
```

### Show edges with Sobel

```python
from skimage.filters import sobel
```

```python
edges_sobel_penguin = sobel(gray_penguin_image)       #filters in skimage must be gray scaled
show_image(edges_sobel_penguin, title='Edges using Sobel filter')
```

### Show edges with Canny

```python
edges_canny_penguin = cv2.Canny(penguin_image,100,100)      #Canny filter is more powerful
show_image(edges_sobel_penguin, title='Edges using Canny filter')
```

# <b>2.  Blurring

### Read and display notebook image

```python
notebook_image = plt.imread('notebook.jpg')
show_image(notebook_image, title='Notebook')
```

### Gaussian blur the image

```python
from skimage.filters import gaussian
```

```python
blurred_notebook = gaussian(notebook_image, 18, channel_axis=-1)    #gaussian filter expects the image is gray, so we should use channel_axis
show_image(blurred_notebook, title='Blurred Apples')
```

```python
blurred_notebook = gaussian(color.rgb2gray(notebook_image), 18)     #gray scaled so we don't need to use channel_axis=-1
show_image(blurred_notebook, title='Blurred Apples')
```

# 3.<b> Contrast Enhancement

### Read and display chest image

```python
chest_img = plt.imread('chest.jpg') 
show_image(chest_img, title='Chest X-ray')
```

```python
# Plot it's histogram
plt.hist(chest_img.ravel());
```

### Applay histogram equalization to enhance the contrast

```python
from skimage import exposure
```

```python
enhanced_apples = exposure.equalize_hist(chest_img)
show_image(enhanced_apples)
```

```python
plt.hist(enhanced_apples.ravel());
```

# 4. Morphology Transformation

### Read and display letter_img

```python
letter_img = plt.imread('letterR.jpg') 
show_image(letter_img, title='Letter R')
```

### Apply Erosion and Dilation

```python
from skimage import morphology
```

```python
eroded_image = morphology.binary_erosion(color.rgb2gray(letter_img))
show_image(eroded_image)
```

```python
eroded_image = morphology.binary_dilation(color.rgb2gray(letter_img))
show_image(eroded_image)
```
