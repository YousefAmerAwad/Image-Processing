```python
import numpy as np
#import matplotlib_inline as plt
#import matplotlib.image as pltimg
import matplotlib.pyplot as plt
import cv2
```

### Define show image function

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
```

### Display penguin_image

```python
penguin_image = plt.imread('penguin.jpg')
show_image(penguin_image, 'Original Penduin Image')
```

### Use Scilit-image to display the gray scale of the penguin image

```python
from skimage import color
```

```python
gray_penguin_image = color.rgb2gray(penguin_image)
show_image(gray_penguin_image, 'Original Penduin Image')
```

### Show edges with Sobel

```python
from skimage.filters import sobel
```

```python
edges_sobel_penguin = sobel(gray_penguin_image)       #filters in skimage must be gray scaled
show_image(edges_sobel_penguin, title='Edges using Sobel filter')
```

### Show edges with Canny

```python
edges_canny_penguin = cv2.Canny(penguin_image,100,100)      #Canny filter is more powerful
show_image(edges_sobel_penguin, title='Edges using Canny filter')
```
