## Import Some Important Libraries

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
import pylab 
pylab.rcParams['figure.figsize'] = [10,8]
```

## Read Image

```python
img = cv2.imread('Lenna_(test_image).png')
```

## Print Image Size

```python
img.size
```

## Print Image Shape

```python
img.shape
```

## Print Image dType

```python
img.dtype
```

## Plot The Image

```python
plt.imshow(img)
```

## Plot a single channel

```python
b, g, r = cv2.split(img)
```

```python
plt.imshow(r)
```

```python
plt.imshow(r, cmap='gray')
```

```python
plt.imshow(b, cmap='gray')
```

## Convert to gray scale image 

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
```

## Fix the awkward thing

```python
fixed_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fixed_img)
```

```python
fixed_img = cv2.merge((r,g,b))
plt.imshow(fixed_img)
```

## Get Lenna's Face

```python
Lenna_face = fixed_img[130:250, 140:230]
plt.imshow(Lenna_face)
```

## Vertical Flip

```python
v_flip = cv2.flip(fixed_img, 0)
plt.imshow(v_flip)
```

## Horizontal Flip

```python
h_flip = cv2.flip(fixed_img, 1)
plt.imshow(h_flip)
```

## Transpose

```python
trans_img = cv2.transpose(fixed_img)
plt.imshow(trans_img)
```

## Blurring (Noise Reduction)

```python
img_blur = cv2.GaussianBlur(fixed_img, (5,5), -1)
plt.imshow(img_blur)
```
